```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy.random import rand
from matplotlib import font_manager,rc


font_name=font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)

df = pd.read_csv('2011년.txt',header=2,sep='\t',thousands = ',')
new_df=df.drop(0)
new_df=new_df[['지역','계','계.1']]
new_df.columns=['구','총인구','65세이상인구']
new_df['65세이상비율'] = new_df.apply(lambda row: (row['65세이상인구'] / row['총인구']),axis=1)
new_df
```

```python
구별노인인구=new_df.pivot_table(index='구',values='65세이상인구')
구별노인인구.plot(kind='barh',figsize=(8,8),grid=True)
```

<h2>2011~2020 서울 지역구별 1분기 65세이상인구/비율</h2>

```python
df = pd.read_csv('2011_2020.서울노인인구(지역구).txt',header=2,sep='\t',thousands = ',')
new_df=df.drop(0)
new_df=new_df[['기간','지역','계','계.1']]
new_df=new_df[new_df['지역']!='합계']
new_df=new_df[new_df['기간'].str.contains('.1/4')]
new_df.columns=['년도','구','총인구','65세이상인구']
new_df['65세이상비율'] = new_df.apply(lambda row: (row['65세이상인구'] / row['총인구']),axis=1)
new_df
```

```python
year=input(' 2011~2020 년도를 숫자만 입력하세요. ')
year_df=new_df[new_df['년도'].str.contains(year)]
년도별노인인구비율=year_df.pivot_table(index='구',values='65세이상비율')
년도별노인인구비율.plot(kind='barh',figsize=(8,8),grid=True,title="%s년도 노인비율"%(year))
```

```python
location=input(' 지역구를 입력하세요. ')
location_df=new_df[new_df['구'].str.contains(location)]
지역구별노인인구비율=location_df.pivot_table(index='년도',values='65세이상비율')
지역구별노인인구비율.plot(kind='barh',figsize=(8,8),grid=True,title="%s"%(location))
```

```python
import pandas as pd
import folium
import matplotlib.pyplot as plt
# 미리 만들어둔 데이터를 불러옵니다.
geo_data = './구.geojson'
center = [37.541, 126.986]
# 맵이 center 에 위치하고, zoom 레벨은 11로 시작하는 맵 m을 만듭니다.
m = folium.Map(location=center, zoom_start=10)
# Choropleth 레이어를 만들고, 맵 m에 추가합니다.
folium.Choropleth( geo_data=geo_data, data=new_df, columns=('구', '65세이상비율'),
 key_on='feature.properties.SIG_KOR_NM',
 fill_color='BuPu',
 legend_name='노령 인구비율', ).add_to(m)
m.save('서울비율.html')
```
